```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
# Load dataset
df = pd.read_csv("salary/Salary_dataset.csv")

# Features and target
X = df[['YearsExperience']]
y = df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse_no_shuffle = root_mean_squared_error(y_test, y_pred)

print("MSE without shuffling:", mse_no_shuffle)
```

```python
mse_shuffled = []

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, shuffle=True, random_state=i
    )

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    mse = root_mean_squared_error(y_test, y_pred)
    mse_shuffled.append(mse)

    print(f"Shuffle {i+1} - MSE: {mse}")
```

```python
print("\nMSE without shuffling:", mse_no_shuffle)
print("MSE with shuffling (5 runs):", mse_shuffled)
print("Average MSE with shuffling:", np.mean(mse_shuffled))
```
